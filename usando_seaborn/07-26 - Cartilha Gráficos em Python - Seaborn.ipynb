```python
import seaborn as sns # importa a biblioteca seaborn e renomeia como sns    
import matplotlib.pyplot as plt # importa a biblioteca matplotlib e renomeia como plt
import pandas as pd # importa a biblioteca pandas e renomeia como pd
```

## Importando Base de Dados

```python
df_videos=pd.read_excel('videosYT.xlsx',"videos") # importa os dados abrindo o excel
display(df_videos) # mostra os dados na tela
```

## Tipos de Gráficos

Documentação : https://seaborn.pydata.org/api.html#distribution-api



## Criando um gráfico de dispersão:
Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df_videos, # cria o gráfico de dispersão
                      x="Nº de Views",
                      y="Nº de Likes",
                      hue="Responsável",
                      style="Responsável",
                      palette=["red","black"],
                      size = "Nº Inscrições",
                     legend="full")
```

## Criando um gráfico de dispersão(relacional):

Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot


```python
fig = sns.relplot(data=df_videos, x="Nº de Views",y="Nº de Likes",col="Categoria")#kind="line"
fig.set_axis_labels(x_var="N° views", y_var="N° Likes") # altera o nome dos eixos
fig.set_titles("Categoria:{col_name}") # altera o titulo do gráfico
plt.show(fig) # exibe o gráfico
```

## Criando um gráfico de linha:
Documentação : https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
df_inscritos=pd.read_excel('videosYT.xlsx',"Inscritos") # importa os dados abrindo o excel
display(df_inscritos) # mostra os dados na tela

graf_linha = sns.lineplot(data = df_inscritos, x="Mês/Ano", y="Inscritos", color = "red" ) # cria o gráfico de linha
graf_linha.set_title("Crescimento Inscritos") # adiciona um titulo ao gráfico
plt.show(graf_linha)   # exibe o gráfico
```

## Criando Histogramas:

Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


```python
import seaborn as sns # importa a biblioteca seaborn e renomeia como sns    
import matplotlib.pyplot as plt # importa a biblioteca matplotlib e renomeia como plt
import pandas as pd # importa a biblioteca pandas e renomeia como pd

df_videos=pd.read_excel('videosYT.xlsx',"videos") # importa os dados abrindo o excel
display(df_videos) # mostra os dados na tela

fig = sns.displot(data = df_videos,  x = "Nº de Likes", hue="Responsável", col="Categoria" )#kind = "kde"
plt.show(fig) # exibe o gráfico
```

```python
fig = sns.pairplot(data=df_videos, hue="Categoria") # cria o gráfico de dispersão
plt.show(fig) # imprime o grafico na tela
```

## Criando uma regressão linear:

Documentação: https://seaborn.pydata.org/tutorial/regression.html#functions-to-draw-linear-regression-models

```python
fig = sns.regplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes") # cria o gráfico de regressão
plt.show(fig)
```

```python
fig = sns.lmplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes", hue= "Responsável", markers = ['o','x']) # cria o gráfico de regressão 
plt.show(fig)
```
